```python
import pandas as pd
import numpy as np
```

```python
data_car=pd.read_csv(r"E:\Users\Documents\College\Engineering\F.E. Eng\ACAIDS\used cars lecture 4.csv")
print(data_car)
```

```python
import seaborn as sns
sns.pairplot(data_car,height=2,aspect=1.5)
```

```python
sns.heatmap(data_car.corr(),annot=True,linewidth=0.6)
```

```python
x=data_car.drop(["KMS","mileage","price","torque","power"],axis="columns")
print(x)
y=data_car.drop(["KMS","mileage","engine","power","torque",],axis="columns")
print(y)
```

```python
from sklearn.model_selection import train_test_split
x1,x2,y1,y2=train_test_split(x,y,test_size=0.3,random_state=2)
print(x1)
print(x2)
print(y1)
print(y2)
```

```python
from sklearn.linear_model import LinearRegression
equation=LinearRegression()
```

```python
equation.fit(x1,y1)
```

```python
print(equation.intercept_)
print(equation.coef_)
```

```python
y_predicted=equation.predict(x2)
print(y_predicted)
```

```python
from sklearn import metrics
MAE=metrics.mean_absolute_error(y2,y_predicted)
MAE
```

```python
Rsquare=metrics.r2_score(y2,y_predicted)
Rsquare
```
